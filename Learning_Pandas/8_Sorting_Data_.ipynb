# Sorting Data

```python
import pandas as pd
```

```python
people = {
    "First_Name": ['Gautam','Soumya','Ashutosh'],
    "Last_Name" : ['Jha','Yadav','Yadav'],
    "Email"     : ['Gautamjha@mail.com','SoumyaYadav@mail.com','AshutoshYadav@mail.com']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

##### using .sort_values to sort a column

```python
df.sort_values(by='First_Name',ascending=False)
```

##### to sort multiple cols

```python
# this will first sort the first name and if there is a common first name then it will sort the last name
df.sort_values(by=['First_Name','Last_Name'],ascending = [False,True])
```

```python
df.sort_index()
```
